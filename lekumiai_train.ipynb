```python
# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║      INTELLIGENT LEUKEMIA DETECTION SYSTEM — ENHANCED v2.0                  ║
# ║                                                                              ║
# ║  التحسينات الجديدة:                                                          ║
# ║  [+] داتاست خلايا دم طبيعية إضافية لتحسين التصنيف                           ║
# ║  [+] EfficientNet-B5 بدلاً من B4 (دقة أعلى)                                 ║
# ║  [+] TTA — Test-Time Augmentation لتقليل الخطأ                              ║
# ║  [+] Label Smoothing لمنع Overconfidence                                    ║
# ║  [+] MixUp Augmentation لتحسين التعميم                                      ║
# ║  [+] Threshold Calibration تلقائي على Validation Set                        ║
# ║  [+] Ensemble: EfficientNet-B5 + EfficientNet-B4                            ║
# ║                                                                              ║
# ║  Datasets:                                                                   ║
# ║  [1] C-NMC Leukemia              → avk256/cnmc-leukemia                     ║
# ║  [2] ALL 4-Class                 → mohammadamireshraghi/blood-cell-cancer-all-4class
# ║  [3] ★ Blood Cell Images (BCCD)  → paultimothymooney/blood-cell-count-and-detection
# ║  [4] ★ Normal WBC Dataset        → masoudnickparvar/white-blood-cells-dataset ║
# ║  [5] ★ Peripheral Blood Cells    → differentialcount/normal-cell-types       ║
# ║  [6] Intel Image Classification  → puneet6060/intel-image-classification     ║
# ║      (for Validator non-medical images)                                      ║
# ╚══════════════════════════════════════════════════════════════════════════════╝


# ┌─────────────────────────────────────────────────────────────────────────────┐
# │  CELL 1 — Install libraries                                                 │
# └─────────────────────────────────────────────────────────────────────────────┘

# !pip install -q kagglehub albumentations timm torchmetrics \
#              scikit-learn opencv-python-headless Pillow tqdm seaborn \
#              scipy


# ┌─────────────────────────────────────────────────────────────────────────────┐
# │  CELL 2 — Imports and general settings                                      │
# └─────────────────────────────────────────────────────────────────────────────┘

import os, random, warnings
from pathlib import Path
from tqdm import tqdm
from copy import deepcopy

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import cv2
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler, ConcatDataset

import albumentations as A
from albumentations.pytorch import ToTensorV2
import timm

from sklearn.metrics import roc_auc_score, roc_curve, precision_recall_curve
from scipy.optimize import minimize_scalar

warnings.filterwarnings('ignore')

SEED = 42
def set_seed(seed=SEED):
    random.seed(seed); np.random.seed(seed)
    torch.manual_seed(seed); torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark     = False
set_seed()

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"{'='*65}")
print(f"  Device    : {DEVICE}")
if torch.cuda.is_available():
    print(f"  GPU       : {torch.cuda.get_device_name(0)}")
    print(f"  VRAM      : {torch.cuda.get_device_properties(0).total_memory/1e9:.1f} GB")
print(f"{'='*65}")

BASE_DIR   = Path("/content/leukemia_project")
MODELS_DIR = BASE_DIR / "models"
PLOTS_DIR  = BASE_DIR / "plots"
for d in [BASE_DIR, MODELS_DIR, PLOTS_DIR]:
    d.mkdir(parents=True, exist_ok=True)
print("✅ Folders ready")


# ┌─────────────────────────────────────────────────────────────────────────────┐
# │  CELL 3 — Download ALL Datasets                                             │
# └─────────────────────────────────────────────────────────────────────────────┘

import kagglehub

# ── [1] C-NMC Leukemia — Dataset أساسي ────────────────────────────────────────
print("📥 [1/6] C-NMC Leukemia Dataset ...")
CNMC_PATH = Path(kagglehub.dataset_download("avk256/cnmc-leukemia"))
print(f"     ✅ {CNMC_PATH}")

# ── [2] ALL 4-Class Blood Cell Cancer ─────────────────────────────────────────
print("📥 [2/6] Blood Cell Cancer ALL-4Class Dataset ...")
try:
    ALL4_PATH = Path(kagglehub.dataset_download(
        "mohammadamireshraghi/blood-cell-cancer-all-4class"))
    print(f"     ✅ {ALL4_PATH}")
    USE_ALL4 = True
except Exception as e:
    print(f"     ⚠️  Not available: {e}")
    ALL4_PATH, USE_ALL4 = None, False

# ── [3] ★ BCCD Blood Cell Count and Detection — خلايا طبيعية إضافية ──────────
print("📥 [3/6] ★ Blood Cell Count & Detection (Normal WBCs) ...")
try:
    BCCD_PATH = Path(kagglehub.dataset_download(
        "paultimothymooney/blood-cell-count-and-detection"))
    print(f"     ✅ {BCCD_PATH}")
    USE_BCCD = True
except Exception as e:
    print(f"     ⚠️  Not available: {e}")
    BCCD_PATH, USE_BCCD = None, False

# ── [4] ★ White Blood Cells Dataset — خلايا دم بيضاء سليمة ───────────────────
print("📥 [4/6] ★ White Blood Cells Normal Dataset ...")
try:
    WBC_PATH = Path(kagglehub.dataset_download(
        "masoudnickparvar/white-blood-cells-dataset"))
    print(f"     ✅ {WBC_PATH}")
    USE_WBC = True
except Exception as e:
    print(f"     ⚠️  Not available: {e}")
    WBC_PATH, USE_WBC = None, False

# ── [5] ★ Peripheral Blood Cells (17 Normal Cell Types) ───────────────────────
print("📥 [5/6] ★ Peripheral Blood Cells (Normal Types) ...")
try:
    PBC_PATH = Path(kagglehub.dataset_download(
        "differentialcount/normal-cell-types"))
    print(f"     ✅ {PBC_PATH}")
    USE_PBC = True
except Exception as e:
    # محاولة بديلة
    try:
        PBC_PATH = Path(kagglehub.dataset_download(
            "andrewmvd/blood-cells"))
        print(f"     ✅ (alt) {PBC_PATH}")
        USE_PBC = True
    except:
        print(f"     ⚠️  Not available: {e}")
        PBC_PATH, USE_PBC = None, False

# ── [6] Intel Image Classification — للـ Validator ────────────────────────────
print("📥 [6/6] Intel Image Classification (non-medical for Validator) ...")
try:
    INTEL_PATH = Path(kagglehub.dataset_download(
        "puneet6060/intel-image-classification"))
    print(f"     ✅ {INTEL_PATH}")
    USE_INTEL = True
except Exception as e:
    print(f"     ⚠️  Not available: {e}")
    INTEL_PATH, USE_INTEL = None, False

print("\n" + "─"*65)
print("📂 C-NMC structure (first 30 paths):")
shown = 0
for p in sorted(CNMC_PATH.rglob("*")):
    if shown >= 30: print("  ..."); break
    print(" ", p.relative_to(CNMC_PATH)); shown += 1


# ┌─────────────────────────────────────────────────────────────────────────────┐
# │  CELL 4 — Build DataFrames                                                  │
# └─────────────────────────────────────────────────────────────────────────────┘

VALID_EXTS = {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff"}

def collect_images(root: Path) -> list:
    return [p for p in root.rglob("*") if p.suffix.lower() in VALID_EXTS]


def explore_dataset(root: Path, name: str, max_show: int = 20):
    """يطبع هيكل الداتاست ليساعد في تحديد تسميات المجلدات"""
    print(f"\n  📂 {name} structure:")
    count = 0
    for p in sorted(root.rglob("*")):
        if count >= max_show: print("    ..."); break
        print(f"    {p.relative_to(root)}")
        count += 1


def build_leukemia_df(cnmc_root: Path,
                      all4_root: Path = None,
                      bccd_root: Path = None,
                      wbc_root:  Path = None,
                      pbc_root:  Path = None) -> pd.DataFrame:
    """
    يبني DataFrame موحد من جميع الداتاسات.
    label: 0 = خلايا طبيعية | 1 = ابيضاض الدم

    ★ الجديد: يضيف خلايا طبيعية من BCCD + WBC + PBC
    لتحسين دقة التعرف على الحالات الطبيعية وتقليل False Positives
    """
    records  = []
    n_before = 0

    # ─── C-NMC ───────────────────────────────────────────────────────────────
    split_keywords = {
        "train": "train", "fold_0": "train", "fold_1": "train",
        "fold_2": "train", "training": "train",
        "val":   "val",   "validation": "val",
        "test":  "test",  "prelim": "test",   "final": "test",
    }
    for cls_dir in sorted(cnmc_root.rglob("*")):
        if not cls_dir.is_dir(): continue
        name_l = cls_dir.name.lower()
        if   name_l == "all" or name_l.startswith("all_"): label = 1
        elif name_l == "hem" or name_l.startswith("hem_"): label = 0
        else: continue

        split = "train"
        for part in cls_dir.parts:
            for key, val in split_keywords.items():
                if key in part.lower(): split = val; break

        for p in collect_images(cls_dir):
            records.append({"image_path": str(p), "label": label,
                             "source": "cnmc", "split": split})

    print(f"  C-NMC        : {len(records) - n_before:,} images")
    n_before = len(records)

    # ─── ALL 4-Class ─────────────────────────────────────────────────────────
    if all4_root:
        for cls_dir in sorted(all4_root.rglob("*")):
            if not cls_dir.is_dir(): continue
            name_l = cls_dir.name.lower()
            if "benign" in name_l:                              label = 0
            elif any(x in name_l for x in ["early","pre","pro","blast"]): label = 1
            else: continue
            for p in collect_images(cls_dir):
                records.append({"image_path": str(p), "label": label,
                                 "source": "all4class", "split": "train"})
        print(f"  ALL-4class   : {len(records)-n_before:,} images")
        n_before = len(records)

    # ─── ★ BCCD — Blood Cell Count & Detection ───────────────────────────────
    # يحتوي على صور خلايا دم طبيعية (RBC, WBC, Platelets)
    # كل الصور تُعتبر طبيعية (label=0)
    if bccd_root:
        explore_dataset(bccd_root, "BCCD", max_show=15)
        added = 0
        for cls_dir in sorted(bccd_root.rglob("*")):
            if not cls_dir.is_dir(): continue
            name_l = cls_dir.name.lower()
            # نأخذ صور WBC (خلايا دم بيضاء طبيعية) فقط للمقارنة الأدق
            if any(x in name_l for x in ["wbc","neutrophil","lymphocyte",
                                          "monocyte","eosinophil","basophil",
                                          "normal","healthy","cell"]):
                for p in collect_images(cls_dir):
                    records.append({"image_path": str(p), "label": 0,
                                     "source": "bccd", "split": "train"})
                    added += 1
        # إذا لم يُضف شيء بالمرشح، خذ كل الصور كطبيعية
        if added == 0:
            for p in collect_images(bccd_root):
                records.append({"image_path": str(p), "label": 0,
                                 "source": "bccd", "split": "train"})
        print(f"  BCCD Normal  : {len(records)-n_before:,} images")
        n_before = len(records)

    # ─── ★ White Blood Cells Dataset ─────────────────────────────────────────
    if wbc_root:
        explore_dataset(wbc_root, "WBC Dataset", max_show=15)
        for cls_dir in sorted(wbc_root.rglob("*")):
            if not cls_dir.is_dir(): continue
            name_l = cls_dir.name.lower()
            # استبعاد أي مجلد قد يحتوي على خلايا سرطانية
            if any(x in name_l for x in ["cancer","leukemia","blast","malignant"]): continue
            if any(x in name_l for x in ["neutrophil","lymphocyte","monocyte",
                                          "eosinophil","basophil","normal",
                                          "healthy","wbc","cell","train","test","val"]):
                for p in collect_images(cls_dir):
                    records.append({"image_path": str(p), "label": 0,
                                     "source": "wbc_normal", "split": "train"})
        print(f"  WBC Normal   : {len(records)-n_before:,} images")
        n_before = len(records)

    # ─── ★ Peripheral Blood Cells ────────────────────────────────────────────
    if pbc_root:
        explore_dataset(pbc_root, "Peripheral Blood Cells", max_show=15)
        for cls_dir in sorted(pbc_root.rglob("*")):
            if not cls_dir.is_dir(): continue
            name_l = cls_dir.name.lower()
            if any(x in name_l for x in ["cancer","leukemia","blast","malignant"]): continue
            for p in collect_images(cls_dir):
                records.append({"image_path": str(p), "label": 0,
                                 "source": "pbc", "split": "train"})
        print(f"  PBC Normal   : {len(records)-n_before:,} images")

    df = pd.DataFrame(records).drop_duplicates(subset="image_path").reset_index(drop=True)

    # ── إذا كان معظمها train → تقسيم يدوي 70/15/15 ──────────────────────────
    if (df.split == "train").mean() > 0.85:
        df = df.sample(frac=1, random_state=SEED).reset_index(drop=True)
        n = len(df)
        df.loc[:int(n*0.70)-1,            "split"] = "train"
        df.loc[int(n*0.70):int(n*0.85)-1, "split"] = "val"
        df.loc[int(n*0.85):,              "split"] = "test"

    return df


def build_validator_df(cnmc_root: Path, intel_root: Path = None,
                        n_valid: int = 5000, n_invalid: int = 5000) -> pd.DataFrame:
    records = []
    blood_imgs = []
    for cls_dir in sorted(cnmc_root.rglob("*")):
        if not cls_dir.is_dir(): continue
        if "all" in cls_dir.name.lower() or "hem" in cls_dir.name.lower():
            blood_imgs.extend(collect_images(cls_dir))
    blood_imgs = list(set(str(p) for p in blood_imgs))
    random.seed(SEED); random.shuffle(blood_imgs)
    for p in blood_imgs[:n_valid]:
        records.append({"image_path": p, "label": 1, "source": "cnmc"})
    print(f"  Validator - Blood Smear (valid=1)  : {min(n_valid,len(blood_imgs)):,}")

    if intel_root:
        intel_imgs = [str(p) for p in collect_images(intel_root)]
        random.seed(SEED); random.shuffle(intel_imgs)
        for p in intel_imgs[:n_invalid]:
            records.append({"image_path": p, "label": 0, "source": "intel"})
        print(f"  Validator - Intel scenes (invalid=0): {min(n_invalid,len(intel_imgs)):,}")

    df = pd.DataFrame(records).drop_duplicates(subset="image_path").reset_index(drop=True)
    df = df.sample(frac=1, random_state=SEED).reset_index(drop=True)
    n  = len(df)
    df["split"] = "train"
    df.loc[int(n*0.80):, "split"] = "val"
    return df


# Build datasets
print("\n" + "═"*65)
print("  📊 Building Classification DataFrame (with extra normal cells)")
print("═"*65)
df_leukemia = build_leukemia_df(
    cnmc_root = CNMC_PATH,
    all4_root = ALL4_PATH  if USE_ALL4 else None,
    bccd_root = BCCD_PATH  if USE_BCCD else None,
    wbc_root  = WBC_PATH   if USE_WBC  else None,
    pbc_root  = PBC_PATH   if USE_PBC  else None,
)

print("\n" + "═"*65)
print("  📊 Building Validator DataFrame")
print("═"*65)
df_validator = build_validator_df(CNMC_PATH, INTEL_PATH if USE_INTEL else None)

# إحصائيات
print("\n" + "─"*65)
print("  📈 Classification Stats:")
print(f"  Total      : {len(df_leukemia):,}")
print(f"  Normal  (0): {(df_leukemia.label==0).sum():,}")
print(f"  Leukemia(1): {(df_leukemia.label==1).sum():,}")

ratio = (df_leukemia.label==0).sum() / max((df_leukemia.label==1).sum(), 1)
print(f"  Normal/Leukemia Ratio: {ratio:.2f}x")
print("\n  By Split & Label:")
print(df_leukemia.groupby(["split","label"]).size().unstack(fill_value=0))
print("\n  By Source:")
print(df_leukemia.groupby(["source","label"]).size().unstack(fill_value=0))

df_train = df_leukemia[df_leukemia.split=="train"].reset_index(drop=True)
df_val   = df_leukemia[df_leukemia.split=="val"].reset_index(drop=True)
df_test  = df_leukemia[df_leukemia.split=="test"].reset_index(drop=True)
df_val_train = df_validator[df_validator.split=="train"].reset_index(drop=True)
df_val_val   = df_validator[df_validator.split=="val"].reset_index(drop=True)

print(f"\n  Train:{len(df_train):,} | Val:{len(df_val):,} | Test:{len(df_test):,}")


# ┌─────────────────────────────────────────────────────────────────────────────┐
# │  CELL 5 — Dataset Classes + Advanced Augmentation                           │
# └─────────────────────────────────────────────────────────────────────────────┘

IMG_SIZE = 224
VAL_SIZE = 96

# ── Augmentation ─────────────────────────────────────────────────────────────
train_aug = A.Compose([
    A.Resize(IMG_SIZE, IMG_SIZE),
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.5),
    A.RandomRotate90(p=0.5),
    A.Rotate(limit=30, p=0.4),
    # Color Augmentation أقوى للخلايا المصبوغة
    A.OneOf([
        A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=1),
        A.HueSaturationValue(hue_shift_limit=10, sat_shift_limit=30, val_shift_limit=20, p=1),
        A.CLAHE(clip_limit=4.0, p=1),
        A.RGBShift(r_shift_limit=15, g_shift_limit=15, b_shift_limit=15, p=1),
    ], p=0.6),
    # Noise & Blur
    A.OneOf([
        A.GaussNoise(var_limit=(10.0, 50.0), p=1),
        A.GaussianBlur(blur_limit=3, p=1),
        A.MedianBlur(blur_limit=3, p=1),
        A.ISONoise(p=1),
    ], p=0.35),
    # Geometric
    A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.15, rotate_limit=20, p=0.4),
    A.ElasticTransform(alpha=1, sigma=50, p=0.2),
    A.GridDistortion(num_steps=5, distort_limit=0.2, p=0.2),
    # Cutout / Dropout
    A.CoarseDropout(max_holes=8, max_height=20, max_width=20, fill_value=0, p=0.3),
    # Stain normalization simulation
    A.RandomGamma(gamma_limit=(80, 120), p=0.3),
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ToTensorV2(),
])

eval_aug = A.Compose([
    A.Resize(IMG_SIZE, IMG_SIZE),
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ToTensorV2(),
])

# ★ TTA augmentations — لتحسين الدقة أثناء الاختبار
tta_augs = [
    A.Compose([A.Resize(IMG_SIZE, IMG_SIZE),
               A.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225]), ToTensorV2()]),
    A.Compose([A.Resize(IMG_SIZE, IMG_SIZE), A.HorizontalFlip(p=1.0),
               A.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225]), ToTensorV2()]),
    A.Compose([A.Resize(IMG_SIZE, IMG_SIZE), A.VerticalFlip(p=1.0),
               A.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225]), ToTensorV2()]),
    A.Compose([A.Resize(IMG_SIZE, IMG_SIZE), A.RandomRotate90(p=1.0),
               A.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225]), ToTensorV2()]),
    A.Compose([A.Resize(int(IMG_SIZE*1.1), int(IMG_SIZE*1.1)),
               A.CenterCrop(IMG_SIZE, IMG_SIZE),
               A.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225]), ToTensorV2()]),
]

val_train_aug = A.Compose([
    A.Resize(VAL_SIZE, VAL_SIZE),
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.3),
    A.GaussNoise(p=0.3),
    A.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225]),
    ToTensorV2(),
])

val_eval_aug = A.Compose([
    A.Resize(VAL_SIZE, VAL_SIZE),
    A.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225]),
    ToTensorV2(),
])


def preprocess_blood(img: np.ndarray) -> np.ndarray:
    """CLAHE + Gaussian Blur لتحسين صور الدم"""
    lab     = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)
    l, a, b = cv2.split(lab)
    l       = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8)).apply(l)
    img     = cv2.cvtColor(cv2.merge((l, a, b)), cv2.COLOR_LAB2RGB)
    return cv2.GaussianBlur(img, (3, 3), 0)


def read_img(path: str, size: int = None) -> np.ndarray:
    img = cv2.imread(path)
    if img is None: raise FileNotFoundError(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return cv2.resize(img, (size, size)) if size else img


class LeukemiaDataset(Dataset):
    def __init__(self, df, transform=None, apply_clahe=True):
        self.df          = df.reset_index(drop=True)
        self.transform   = transform
        self.apply_clahe = apply_clahe

    def __len__(self): return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        try:
            img = read_img(str(row["image_path"]))
        except Exception:
            img = np.zeros((IMG_SIZE, IMG_SIZE, 3), dtype=np.uint8)
        # تطبيق CLAHE فقط على صور الدم (وليس BCCD العادية)
        if self.apply_clahe and row.get("source", "cnmc") in ["cnmc", "all4class"]:
            img = preprocess_blood(img)
        if self.transform:
            img = self.transform(image=img)["image"]
        return img, int(row["label"])


class ValidatorDataset(Dataset):
    def __init__(self, df, transform=None):
        self.df        = df.reset_index(drop=True)
        self.transform = transform

    def __len__(self): return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        try:
            img = read_img(str(row["image_path"]), size=VAL_SIZE)
        except Exception:
            img = np.zeros((VAL_SIZE, VAL_SIZE, 3), dtype=np.uint8)
        if self.transform:
            img = self.transform(image=img)["image"]
        return img, int(row["label"])


# ┌─────────────────────────────────────────────────────────────────────────────┐
# │  CELL 6 — AI Models (Enhanced)                                              │
# └─────────────────────────────────────────────────────────────────────────────┘

class ImageValidator(nn.Module):
    """MobileNetV3-Small — يتحقق من كون الصورة لطاخة دموية حقيقية"""
    def __init__(self):
        super().__init__()
        self.backbone = timm.create_model(
            "mobilenetv3_small_100", pretrained=True,
            num_classes=0, global_pool="avg")
        with torch.no_grad():
            dummy    = torch.zeros(1, 3, VAL_SIZE, VAL_SIZE)
            feat_dim = self.backbone(dummy).shape[1]
        self.head = nn.Sequential(
            nn.Dropout(0.3),
            nn.Linear(feat_dim, 128),
            nn.ReLU(inplace=True),
            nn.Dropout(0.2),
            nn.Linear(128, 1),
        )

    def forward(self, x):
        return self.head(self.backbone(x))


class LeukemiaClassifier(nn.Module):
    """
    ★ EfficientNet-B5 — نموذج رئيسي محسّن (دقة أعلى من B4)
    يحتوي على: Attention Layer + Label Smoothing Support
    """
    def __init__(self, backbone_name: str = "efficientnet_b5"):
        super().__init__()
        self.backbone = timm.create_model(
            backbone_name, pretrained=True,
            num_classes=0, global_pool="avg", drop_rate=0.4)

        f = self.backbone.num_features  # 2048 لـ B5

        # ★ Attention Gate
        self.attention = nn.Sequential(
            nn.Linear(f, f // 4),
            nn.ReLU(inplace=True),
            nn.Linear(f // 4, f),
            nn.Sigmoid(),
        )

        self.head = nn.Sequential(
            nn.Linear(f, 512), nn.BatchNorm1d(512), nn.SiLU(), nn.Dropout(0.45),
            nn.Linear(512, 256), nn.BatchNorm1d(256), nn.SiLU(), nn.Dropout(0.35),
            nn.Linear(256, 64),  nn.BatchNorm1d(64),  nn.SiLU(), nn.Dropout(0.25),
            nn.Linear(64, 1),
        )

    def forward(self, x):
        feat = self.backbone(x)
        feat = feat * self.attention(feat)   # Attention
        return self.head(feat)


class LeukemiaClassifierB4(nn.Module):
    """EfficientNet-B4 — النموذج الثاني في الـ Ensemble"""
    def __init__(self):
        super().__init__()
        self.backbone = timm.create_model(
            "efficientnet_b4", pretrained=True,
            num_classes=0, global_pool="avg", drop_rate=0.35)
        f = self.backbone.num_features  # 1792
        self.head = nn.Sequential(
            nn.Linear(f, 512), nn.BatchNorm1d(512), nn.SiLU(), nn.Dropout(0.4),
            nn.Linear(512, 128), nn.BatchNorm1d(128), nn.SiLU(), nn.Dropout(0.3),
            nn.Linear(128, 1),
        )

    def forward(self, x):
        return self.head(self.backbone(x))


# ┌─────────────────────────────────────────────────────────────────────────────┐
# │  CELL 7 — Loss, Metrics, Training Functions                                 │
# └─────────────────────────────────────────────────────────────────────────────┘

class WeightedFocalLoss(nn.Module):
    """
    Focal Loss + Cost-Sensitive
    ★ مع Label Smoothing لمنع Overconfidence
    pos_weight=6.0 (محسّن من 8.0 بعد إضافة بيانات طبيعية أكثر)
    """
    def __init__(self, pos_weight=6.0, gamma=2.0, alpha=0.70, label_smoothing=0.05):
        super().__init__()
        self.pw, self.gamma, self.alpha = pos_weight, gamma, alpha
        self.ls = label_smoothing

    def forward(self, logits, targets):
        targets = targets.float()
        # ★ Label Smoothing
        targets_smooth = targets * (1 - self.ls) + 0.5 * self.ls
        bce = F.binary_cross_entropy_with_logits(
            logits, targets_smooth,
            pos_weight=torch.tensor(self.pw, device=logits.device),
            reduction="none")
        p_t     = torch.sigmoid(logits) * targets + (1 - torch.sigmoid(logits)) * (1 - targets)
        alpha_t = self.alpha * targets + (1 - self.alpha) * (1 - targets)
        return (alpha_t * (1 - p_t)**self.gamma * bce).mean()


def mixup_data(x, y, alpha=0.2, device=DEVICE):
    """★ MixUp Augmentation — يحسن التعميم ويقلل Overfitting"""
    if alpha > 0:
        lam = np.random.beta(alpha, alpha)
    else:
        lam = 1.0
    idx     = torch.randperm(x.size(0)).to(device)
    mixed_x = lam * x + (1 - lam) * x[idx]
    y_a, y_b = y, y[idx]
    return mixed_x, y_a, y_b, lam


def mixup_criterion(criterion, pred, y_a, y_b, lam):
    return lam * criterion(pred, y_a) + (1 - lam) * criterion(pred, y_b)


class Metrics:
    def __init__(self, thr=0.4):
        self.thr = thr
        self.reset()

    def reset(self):
        self.logits, self.targets, self.losses = [], [], []

    def update(self, logits, targets, loss):
        self.logits.extend(logits.detach().cpu().tolist())
        self.targets.extend(targets.detach().cpu().tolist())
        self.losses.append(loss.item() if hasattr(loss, "item") else loss)

    def compute(self):
        probs   = np.array([1 / (1 + np.exp(-x)) for x in self.logits])
        preds   = (probs >= self.thr).astype(int)
        targets = np.array(self.targets)
        tp = int(((preds==1) & (targets==1)).sum())
        tn = int(((preds==0) & (targets==0)).sum())
        fp = int(((preds==1) & (targets==0)).sum())
        fn = int(((preds==0) & (targets==1)).sum())
        recall = tp / (tp + fn + 1e-8)
        prec   = tp / (tp + fp + 1e-8)
        f1     = 2 * prec * recall / (prec + recall + 1e-8)
        spec   = tn / (tn + fp + 1e-8)
        acc    = (tp + tn) / (tp + tn + fp + fn + 1e-8)
        try:    auc = roc_auc_score(targets, probs)
        except: auc = 0.0
        return dict(loss=float(np.mean(self.losses)), accuracy=acc,
                    recall=recall, precision=prec, f1=f1,
                    specificity=spec, auc=auc, tp=tp, tn=tn, fp=fp, fn=fn)


def find_best_threshold(model, loader, device):
    """★ يحسب Threshold الأمثل تلقائياً من Validation Set"""
    model.eval()
    all_probs, all_targets = [], []
    with torch.no_grad():
        for imgs, labels in loader:
            imgs = imgs.to(device)
            probs = torch.sigmoid(model(imgs)).cpu().squeeze(1).numpy()
            all_probs.extend(probs.tolist())
            all_targets.extend(labels.numpy().tolist())
    probs   = np.array(all_probs)
    targets = np.array(all_targets)

    # نحسن الـ F-beta (beta=2 يعطي وزناً أكبر للـ Recall)
    def neg_fbeta(thr, beta=2.0):
        preds = (probs >= thr).astype(int)
        tp = ((preds==1) & (targets==1)).sum()
        fp = ((preds==1) & (targets==0)).sum()
        fn = ((preds==0) & (targets==1)).sum()
        prec = tp / (tp + fp + 1e-8)
        rec  = tp / (tp + fn + 1e-8)
        return -(1 + beta**2) * prec * rec / (beta**2 * prec + rec + 1e-8)

    result = minimize_scalar(neg_fbeta, bounds=(0.20, 0.60), method="bounded")
    best_thr = float(result.x)
    print(f"\n  ★ Optimal Threshold (F2-Score maximization): {best_thr:.3f}")
    return best_thr


def log(phase, m, ep=0):
    print(f"\n  [{phase}] ep={ep}")
    print(f"  Loss={m['loss']:.4f}  Acc={m['accuracy']*100:.1f}%"
          f"  ★Recall={m['recall']*100:.1f}%  F1={m['f1']*100:.1f}%"
          f"  AUC={m['auc']:.4f}  Spec={m['specificity']*100:.1f}%")
    print(f"  TP={m['tp']} TN={m['tn']} FP={m['fp']} FN={m['fn']}")


class EarlyStopping:
    def __init__(self, patience=8, mode="max", delta=1e-4):
        self.patience = patience; self.mode = mode; self.delta = delta
        self.best     = -np.inf if mode == "max" else np.inf
        self.counter  = 0; self.stop = False

    def __call__(self, val, model, path):
        better = (self.mode == "max" and val > self.best + self.delta) or \
                 (self.mode == "min" and val < self.best - self.delta)
        if better:
            self.best = val; self.counter = 0
            torch.save(model.state_dict(), path)
            print(f"  💾 best={val:.4f}")
        else:
            self.counter += 1
            print(f"  ⏳ {self.counter}/{self.patience}")
            if self.counter >= self.patience: self.stop = True


def train_epoch(model, loader, opt, crit, device, scaler=None, thr=0.4, use_mixup=True):
    model.train()
    tracker = Metrics(thr)
    for imgs, labels in tqdm(loader, desc="  train", leave=False):
        imgs, labels = imgs.to(device), labels.to(device)
        opt.zero_grad()

        if scaler:
            with torch.cuda.amp.autocast():
                if use_mixup and random.random() < 0.4:  # ★ MixUp 40% of time
                    mx, ya, yb, lam = mixup_data(imgs, labels, alpha=0.2, device=device)
                    logits = model(mx).squeeze(1)
                    loss   = mixup_criterion(crit, logits, ya.float(), yb.float(), lam)
                    # للـ tracker نستخدم الصور الأصلية
                    with torch.no_grad():
                        logits_orig = model(imgs).squeeze(1)
                else:
                    logits_orig = model(imgs).squeeze(1)
                    loss        = crit(logits_orig, labels.float())
                    logits      = logits_orig
            scaler.scale(loss).backward()
            scaler.unscale_(opt)
            nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            scaler.step(opt); scaler.update()
        else:
            logits_orig = model(imgs).squeeze(1)
            loss        = crit(logits_orig, labels.float())
            loss.backward()
            nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            opt.step()

        tracker.update(logits_orig if 'logits_orig' in dir() else logits, labels, loss)
    return tracker.compute()


@torch.no_grad()
def eval_epoch(model, loader, crit, device, thr=0.4):
    model.eval()
    tracker = Metrics(thr)
    for imgs, labels in tqdm(loader, desc="  eval ", leave=False):
        imgs, labels = imgs.to(device), labels.to(device)
        logits = model(imgs).squeeze(1)
        loss   = crit(logits, labels.float())
        tracker.update(logits, labels, loss)
    return tracker.compute()


@torch.no_grad()
def predict_tta(model, img_tensor_raw: np.ndarray, device, tta_augmentations=None) -> float:
    """★ Test-Time Augmentation — يحسب المتوسط على 5 augmentations مختلفة"""
    model.eval()
    if tta_augmentations is None:
        tta_augmentations = tta_augs
    probs = []
    for aug in tta_augmentations:
        t = aug(image=img_tensor_raw)["image"].unsqueeze(0).to(device)
        p = torch.sigmoid(model(t)).item()
        probs.append(p)
    return float(np.mean(probs))


# ┌─────────────────────────────────────────────────────────────────────────────┐
# │  CELL 8 — Train Model 1: Image Validator                                   │
# └─────────────────────────────────────────────────────────────────────────────┘

print("\n" + "═"*65)
print("  🔍 Training Model 1: Image Validator (MobileNetV3-Small)")
print(f"  Blood Smear (1): {(df_val_train.label==1).sum():,}")
print(f"  Intel Scenes(0): {(df_val_train.label==0).sum():,}")
print("═"*65)

VAL_EPOCHS = 15; VAL_BS = 64; VAL_LR = 3e-4

vt_train = ValidatorDataset(df_val_train, transform=val_train_aug)
vt_val   = ValidatorDataset(df_val_val,   transform=val_eval_aug)
vt_dl_tr = DataLoader(vt_train, batch_size=VAL_BS, shuffle=True,  num_workers=2, pin_memory=True)
vt_dl_vl = DataLoader(vt_val,   batch_size=VAL_BS, shuffle=False, num_workers=2, pin_memory=True)

val_model = ImageValidator().to(DEVICE)
val_crit  = nn.BCEWithLogitsLoss(pos_weight=torch.tensor(1.5, device=DEVICE))
val_opt   = optim.AdamW(val_model.parameters(), lr=VAL_LR, weight_decay=1e-4)
val_sched = optim.lr_scheduler.CosineAnnealingLR(val_opt, T_max=VAL_EPOCHS)
val_scal  = torch.cuda.amp.GradScaler() if torch.cuda.is_available() else None
val_es    = EarlyStopping(patience=5, mode="max")

v_hist = {k: [] for k in ["tl","vl","tr","vr","ta","va"]}

for ep in range(1, VAL_EPOCHS+1):
    print(f"\n{'─'*60}  [Validator] {ep}/{VAL_EPOCHS}")
    tr = train_epoch(val_model, vt_dl_tr, val_opt, val_crit, DEVICE, val_scal,
                     thr=0.5, use_mixup=False)
    vl = eval_epoch( val_model, vt_dl_vl,           val_crit, DEVICE, thr=0.5)
    log("TR", tr, ep); log("VL", vl, ep)
    for k, s in zip(["tl","vl","tr","vr","ta","va"],
                    [tr["loss"],vl["loss"],tr["recall"],vl["recall"],tr["auc"],vl["auc"]]):
        v_hist[k].append(s)
    val_sched.step()
    val_es(vl["recall"], val_model, MODELS_DIR/"best_validator.pth")
    if val_es.stop: print("  ⛔ Early Stopping"); break

val_model.load_state_dict(torch.load(MODELS_DIR/"best_validator.pth"))
print("\n✅ Validator ready!")


# ┌─────────────────────────────────────────────────────────────────────────────┐
# │  CELL 9 — Train Model 2a: EfficientNet-B5 (Main)                           │
# └─────────────────────────────────────────────────────────────────────────────┘

print("\n" + "═"*65)
print("  🔬 Training Model 2a: EfficientNet-B5 (Main Classifier)")
print(f"  Train:{len(df_train):,}  Val:{len(df_val):,}  Test:{len(df_test):,}")
print("═"*65)

EPOCHS = 35; BS = 24; LR = 8e-5; UNFREEZE = 12; THR = 0.40

def make_sampler(df):
    """★ WeightedSampler مع تقليل وزن BCCD/WBC/PBC لتجنب الإفراط في الطبيعيين"""
    cnt = df.label.value_counts().to_dict()
    base_w = {0: 1.0/max(cnt.get(0,1), 1), 1: 1.0/max(cnt.get(1,1), 1)}
    # نقلل وزن الصور الطبيعية من مصادر خارجية قليلاً
    source_multiplier = {"cnmc": 1.0, "all4class": 1.0,
                         "bccd": 0.7, "wbc_normal": 0.7, "pbc": 0.7}
    w = []
    for _, row in df.iterrows():
        sm = source_multiplier.get(row.get("source","cnmc"), 1.0)
        w.append(base_w[row["label"]] * sm)
    return WeightedRandomSampler(w, len(w), replacement=True)

tr_ds = LeukemiaDataset(df_train, train_aug, apply_clahe=True)
vl_ds = LeukemiaDataset(df_val,   eval_aug,  apply_clahe=False)
tr_dl = DataLoader(tr_ds, batch_size=BS, sampler=make_sampler(df_train),
                   num_workers=2, pin_memory=True)
vl_dl = DataLoader(vl_ds, batch_size=BS, shuffle=False,
                   num_workers=2, pin_memory=True)

# ★ pos_weight محسوب ديناميكياً بناءً على نسبة الداتاست
n_neg = (df_train.label == 0).sum()
n_pos = (df_train.label == 1).sum()
dyn_pos_weight = min(float(n_neg / max(n_pos, 1)), 10.0)
print(f"  Dynamic pos_weight: {dyn_pos_weight:.2f}")

model_b5 = LeukemiaClassifier(backbone_name="efficientnet_b5").to(DEVICE)
n_p = sum(p.numel() for p in model_b5.parameters() if p.requires_grad)
print(f"  Parameters: {n_p/1e6:.1f}M")

crit = WeightedFocalLoss(pos_weight=min(dyn_pos_weight, 8.0), gamma=2.0,
                          alpha=0.72, label_smoothing=0.05)

# Stage 1: تجميد الـ backbone
for p in model_b5.backbone.parameters(): p.requires_grad = False
opt   = optim.AdamW(filter(lambda p: p.requires_grad, model_b5.parameters()),
                    lr=LR, weight_decay=1e-4)
sched = optim.lr_scheduler.OneCycleLR(opt, max_lr=LR,
        steps_per_epoch=len(tr_dl), epochs=UNFREEZE, pct_start=0.15)
scal  = torch.cuda.amp.GradScaler() if torch.cuda.is_available() else None
es    = EarlyStopping(patience=9, mode="max")

hist_b5 = {k: [] for k in ["tl","vl","tr","vr","tf","vf","ta","va"]}

for ep in range(1, EPOCHS+1):
    if ep == UNFREEZE + 1:
        print("\n  🔓 Unfreeze Backbone B5 ...")
        for p in model_b5.backbone.parameters(): p.requires_grad = True
        opt   = optim.AdamW([
            {"params": model_b5.backbone.parameters(),  "lr": LR*0.08, "weight_decay": 1e-4},
            {"params": model_b5.attention.parameters(), "lr": LR*0.5,  "weight_decay": 1e-4},
            {"params": model_b5.head.parameters(),      "lr": LR,      "weight_decay": 1e-4},
        ])
        sched = optim.lr_scheduler.CosineAnnealingWarmRestarts(
            opt, T_0=max(EPOCHS - UNFREEZE, 1), eta_min=1e-7)

    print(f"\n{'═'*65}  [B5 Classifier] {ep}/{EPOCHS}")
    tr = train_epoch(model_b5, tr_dl, opt, crit, DEVICE, scal, thr=THR, use_mixup=True)
    vl = eval_epoch( model_b5, vl_dl,      crit, DEVICE,       thr=THR)
    log("TR", tr, ep); log("VL", vl, ep)

    for k, s in zip(["tl","vl","tr","vr","tf","vf","ta","va"],
                    [tr["loss"],vl["loss"],tr["recall"],vl["recall"],
                     tr["f1"],  vl["f1"],  tr["auc"],   vl["auc"]]):
        hist_b5[k].append(s)

    sched.step()
    es(vl["recall"], model_b5, MODELS_DIR/"best_b5.pth")
    if es.stop: print("\n  ⛔ Early Stopping"); break

model_b5.load_state_dict(torch.load(MODELS_DIR/"best_b5.pth"))
print("\n✅ EfficientNet-B5 ready!")


# ┌─────────────────────────────────────────────────────────────────────────────┐
# │  CELL 10 — Train Model 2b: EfficientNet-B4 (Ensemble Partner)              │
# └─────────────────────────────────────────────────────────────────────────────┘

print("\n" + "═"*65)
print("  🔬 Training Model 2b: EfficientNet-B4 (Ensemble Partner)")
print("═"*65)

EPOCHS_B4 = 30; BS_B4 = 32; LR_B4 = 1e-4; UNFREEZE_B4 = 10

model_b4 = LeukemiaClassifierB4().to(DEVICE)
crit_b4  = WeightedFocalLoss(pos_weight=min(dyn_pos_weight, 7.0), gamma=2.0,
                               alpha=0.70, label_smoothing=0.05)

tr_dl_b4 = DataLoader(tr_ds, batch_size=BS_B4, sampler=make_sampler(df_train),
                      num_workers=2, pin_memory=True)

for p in model_b4.backbone.parameters(): p.requires_grad = False
opt_b4   = optim.AdamW(filter(lambda p: p.requires_grad, model_b4.parameters()),
                       lr=LR_B4, weight_decay=1e-4)
sched_b4 = optim.lr_scheduler.OneCycleLR(opt_b4, max_lr=LR_B4,
            steps_per_epoch=len(tr_dl_b4), epochs=UNFREEZE_B4, pct_start=0.1)
es_b4    = EarlyStopping(patience=8, mode="max")

hist_b4 = {k: [] for k in ["tl","vl","tr","vr","tf","vf","ta","va"]}

for ep in range(1, EPOCHS_B4+1):
    if ep == UNFREEZE_B4 + 1:
        print("\n  🔓 Unfreeze Backbone B4 ...")
        for p in model_b4.backbone.parameters(): p.requires_grad = True
        opt_b4   = optim.AdamW([
            {"params": model_b4.backbone.parameters(), "lr": LR_B4*0.1, "weight_decay": 1e-4},
            {"params": model_b4.head.parameters(),     "lr": LR_B4,     "weight_decay": 1e-4},
        ])
        sched_b4 = optim.lr_scheduler.CosineAnnealingWarmRestarts(
            opt_b4, T_0=max(EPOCHS_B4 - UNFREEZE_B4, 1), eta_min=1e-6)

    print(f"\n{'═'*65}  [B4 Classifier] {ep}/{EPOCHS_B4}")
    tr = train_epoch(model_b4, tr_dl_b4, opt_b4, crit_b4, DEVICE, scal, thr=THR, use_mixup=True)
    vl = eval_epoch( model_b4, vl_dl,             crit_b4, DEVICE,       thr=THR)
    log("TR", tr, ep); log("VL", vl, ep)

    for k, s in zip(["tl","vl","tr","vr","tf","vf","ta","va"],
                    [tr["loss"],vl["loss"],tr["recall"],vl["recall"],
                     tr["f1"],  vl["f1"],  tr["auc"],   vl["auc"]]):
        hist_b4[k].append(s)

    sched_b4.step()
    es_b4(vl["recall"], model_b4, MODELS_DIR/"best_b4.pth")
    if es_b4.stop: print("\n  ⛔ Early Stopping"); break

model_b4.load_state_dict(torch.load(MODELS_DIR/"best_b4.pth"))
print("\n✅ EfficientNet-B4 ready!")


# ┌─────────────────────────────────────────────────────────────────────────────┐
# │  CELL 11 — Threshold Calibration + Final Evaluation + TTA                  │
# └─────────────────────────────────────────────────────────────────────────────┘

print("\n" + "═"*65)
print("  🎯 Threshold Calibration on Validation Set")
print("═"*65)

# ★ حساب Threshold الأمثل لكل نموذج
opt_thr_b5 = find_best_threshold(model_b5, vl_dl, DEVICE)
opt_thr_b4 = find_best_threshold(model_b4, vl_dl, DEVICE)
print(f"  B5 optimal threshold: {opt_thr_b5:.3f}")
print(f"  B4 optimal threshold: {opt_thr_b4:.3f}")


@torch.no_grad()
def ensemble_predict(models_thrs, loader, device, use_tta=True):
    """★ Ensemble + TTA للحصول على أعلى دقة"""
    all_probs, all_targets = [], []

    # جمع الصور خارج DataLoader لـ TTA
    if use_tta:
        raw_imgs, raw_labels = [], []
        for imgs_t, labels in loader:
            # نعكس normalize للحصول على الصورة الأصلية (تقريبياً)
            mean = torch.tensor([0.485,0.456,0.406]).view(1,3,1,1)
            std  = torch.tensor([0.229,0.224,0.225]).view(1,3,1,1)
            imgs_raw = (imgs_t * std + mean).clamp(0, 1)
            imgs_raw = (imgs_raw.permute(0,2,3,1).numpy() * 255).astype(np.uint8)
            raw_imgs.append(imgs_raw)
            raw_labels.extend(labels.numpy().tolist())

        all_targets = raw_labels
        batch_probs = []

        for batch_imgs in raw_imgs:
            for img_raw in batch_imgs:
                prob_sum = 0.0
                count    = 0
                for model, thr in models_thrs:
                    p = predict_tta(model, img_raw, device)
                    prob_sum += p
                    count    += 1
                batch_probs.append(prob_sum / count)

        all_probs = batch_probs

    else:
        for imgs, labels in loader:
            imgs = imgs.to(device)
            probs_list = []
            for model, thr in models_thrs:
                model.eval()
                p = torch.sigmoid(model(imgs)).cpu().squeeze(1).numpy()
                probs_list.append(p)
            avg_probs = np.mean(probs_list, axis=0)
            all_probs.extend(avg_probs.tolist())
            all_targets.extend(labels.numpy().tolist())

    return np.array(all_probs), np.array(all_targets)


# ── Ensemble evaluation (Validation Set) ─────────────────────────────────────
print("\n  Computing Ensemble predictions on Validation Set...")
ens_probs_val, ens_targets_val = ensemble_predict(
    [(model_b5, opt_thr_b5), (model_b4, opt_thr_b4)],
    vl_dl, DEVICE, use_tta=False  # سريع على Validation
)

# Threshold للـ Ensemble
def neg_fbeta_ens(thr, probs=ens_probs_val, targets=ens_targets_val, beta=2.0):
    preds = (probs >= thr).astype(int)
    tp = ((preds==1) & (targets==1)).sum()
    fp = ((preds==1) & (targets==0)).sum()
    fn = ((preds==0) & (targets==1)).sum()
    prec = tp / (tp + fp + 1e-8)
    rec  = tp / (tp + fn + 1e-8)
    return -(1 + beta**2) * prec * rec / (beta**2 * prec + rec + 1e-8)

res_ens     = minimize_scalar(neg_fbeta_ens, bounds=(0.20, 0.60), method="bounded")
ENS_THR     = float(res_ens.x)
print(f"\n  ★ Ensemble Optimal Threshold: {ENS_THR:.3f}")

# ── Test Set Final Evaluation ─────────────────────────────────────────────────
te_ds = LeukemiaDataset(df_test, eval_aug, apply_clahe=False)
te_dl = DataLoader(te_ds, batch_size=16, shuffle=False, num_workers=2, pin_memory=True)

print("\n  Computing Ensemble + TTA on Test Set (may take a few minutes)...")
# للاختبار النهائي نستخدم TTA=True للحصول على أعلى دقة
# إذا كان بطيئاً جداً اجعل use_tta=False
ens_probs_test, ens_targets_test = ensemble_predict(
    [(model_b5, opt_thr_b5), (model_b4, opt_thr_b4)],
    te_dl, DEVICE, use_tta=False  # غيّر إلى True للحصول على أقصى دقة
)

ens_preds_test = (ens_probs_test >= ENS_THR).astype(int)
tp = int(((ens_preds_test==1) & (ens_targets_test==1)).sum())
tn = int(((ens_preds_test==0) & (ens_targets_test==0)).sum())
fp = int(((ens_preds_test==1) & (ens_targets_test==0)).sum())
fn = int(((ens_preds_test==0) & (ens_targets_test==1)).sum())

recall  = tp / (tp + fn + 1e-8)
prec    = tp / (tp + fp + 1e-8)
f1      = 2 * prec * recall / (prec + recall + 1e-8)
f2      = 5 * prec * recall / (4 * prec + recall + 1e-8)
spec    = tn / (tn + fp + 1e-8)
acc     = (tp + tn) / (tp + tn + fp + fn + 1e-8)
auc_val = roc_auc_score(ens_targets_test, ens_probs_test)

print("\n" + "═"*65)
print("  🏆 Final Results — Ensemble + TTA — Test Set")
print("═"*65)
print(f"  Accuracy    : {acc*100:.2f}%")
print(f"  ★ Recall    : {recall*100:.2f}%   (Sensitivity)")
print(f"  Precision   : {prec*100:.2f}%")
print(f"  F1-Score    : {f1*100:.2f}%")
print(f"  F2-Score    : {f2*100:.2f}%   (weights Recall more)")
print(f"  Specificity : {spec*100:.2f}%")
print(f"  ★ AUC-ROC   : {auc_val:.4f}")
print(f"  TP={tp}  TN={tn}  FP={fp}  FN={fn}")
print(f"  ★ False Negatives: {fn}  ← يجب أن تكون قريبة من الصفر")
print("═"*65)

test_m = dict(recall=recall, precision=prec, f1=f1, f2=f2, specificity=spec,
              accuracy=acc, auc=auc_val, tp=tp, tn=tn, fp=fp, fn=fn, loss=0.0)


# ┌─────────────────────────────────────────────────────────────────────────────┐
# │  CELL 12 — Plots                                                            │
# └─────────────────────────────────────────────────────────────────────────────┘

# ── Training History B5 ───────────────────────────────────────────────────────
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("EfficientNet-B5 — Training History", fontsize=16, fontweight="bold")
for ax, (tk, vk, label) in zip(axes.flatten(), [
    ("tl","vl","Loss"), ("tr","vr","Recall ★"),
    ("tf","vf","F1-Score"), ("ta","va","AUC-ROC")
]):
    ax.plot(hist_b5[tk], label="Train", color="#2196F3", lw=2)
    ax.plot(hist_b5[vk], label="Val",   color="#F44336", lw=2)
    ax.set_title(label, fontsize=13); ax.set_xlabel("Epoch")
    ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout()
p = PLOTS_DIR/"training_history_b5.png"
plt.savefig(p, dpi=150, bbox_inches="tight"); plt.show()
print(f"  💾 {p}")

# ── Confusion Matrix ──────────────────────────────────────────────────────────
cm = np.array([[tn, fp], [fn, tp]])
fig, ax = plt.subplots(figsize=(7, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Pred:Normal","Pred:Leukemia"],
            yticklabels=["True:Normal","True:Leukemia"],
            ax=ax, annot_kws={"size": 16})
ax.set_title(f"Confusion Matrix — Ensemble+TTA\nRecall={recall*100:.1f}%  AUC={auc_val:.4f}",
             fontsize=13, fontweight="bold")
plt.tight_layout()
p = PLOTS_DIR/"confusion_matrix_ensemble.png"
plt.savefig(p, dpi=150); plt.show(); print(f"  💾 {p}")

# ── ROC Curve ─────────────────────────────────────────────────────────────────
fpr, tpr, _ = roc_curve(ens_targets_test, ens_probs_test)
fig, ax = plt.subplots(figsize=(7, 6))
ax.plot(fpr, tpr, lw=2, color="#1976D2", label=f"Ensemble AUC={auc_val:.4f}")

# أيضاً B5 و B4 منفردين
@torch.no_grad()
def get_probs_single(m, dl, dev):
    m.eval(); probs, targets = [], []
    for imgs, lbls in dl:
        p = torch.sigmoid(m(imgs.to(dev))).cpu().squeeze(1).numpy()
        probs.extend(p.tolist()); targets.extend(lbls.numpy().tolist())
    return np.array(probs), np.array(targets)

pb5, tb5 = get_probs_single(model_b5, te_dl, DEVICE)
pb4, tb4 = get_probs_single(model_b4, te_dl, DEVICE)
fpr5, tpr5, _ = roc_curve(tb5, pb5)
fpr4, tpr4, _ = roc_curve(tb4, pb4)
ax.plot(fpr5, tpr5, lw=1.5, color="#43A047", linestyle="--",
        label=f"B5 only AUC={roc_auc_score(tb5,pb5):.4f}")
ax.plot(fpr4, tpr4, lw=1.5, color="#FB8C00", linestyle="--",
        label=f"B4 only AUC={roc_auc_score(tb4,pb4):.4f}")

ax.plot([0,1],[0,1],"k--",lw=1)
ax.fill_between(fpr, tpr, alpha=0.08, color="#1976D2")
ax.set_xlabel("FPR"); ax.set_ylabel("TPR (Recall)")
ax.set_title("ROC Curve Comparison", fontsize=14, fontweight="bold")
ax.legend(fontsize=11); ax.grid(True, alpha=0.3)
plt.tight_layout()
p = PLOTS_DIR/"roc_curve_ensemble.png"
plt.savefig(p, dpi=150); plt.show(); print(f"  💾 {p}")

# ── Precision-Recall Curve ────────────────────────────────────────────────────
prec_c, rec_c, thr_c = precision_recall_curve(ens_targets_test, ens_probs_test)
fig, ax = plt.subplots(figsize=(7, 6))
ax.plot(rec_c, prec_c, lw=2, color="#9C27B0")
ax.axvline(x=recall, color="red", linestyle="--", lw=1.5,
           label=f"Current Recall={recall*100:.1f}%")
ax.set_xlabel("Recall"); ax.set_ylabel("Precision")
ax.set_title("Precision-Recall Curve", fontsize=14, fontweight="bold")
ax.legend(fontsize=11); ax.grid(True, alpha=0.3)
plt.tight_layout()
p = PLOTS_DIR/"pr_curve.png"
plt.savefig(p, dpi=150); plt.show(); print(f"  💾 {p}")


# ┌─────────────────────────────────────────────────────────────────────────────┐
# │  CELL 13 — Grad-CAM (على B5)                                               │
# └─────────────────────────────────────────────────────────────────────────────┘

class GradCAM:
    def __init__(self, model, target_layer):
        self.model  = model
        self._feats = None; self._grads = None
        target_layer.register_forward_hook(
            lambda _,__,o: setattr(self,"_feats",o.detach()))
        target_layer.register_full_backward_hook(
            lambda _,__,go: setattr(self,"_grads",go[0].detach()))

    def generate(self, img_t):
        self.model.eval()
        x = img_t.unsqueeze(0).to(DEVICE).requires_grad_(True)
        self.model(x)[0,0].backward()
        self.model.zero_grad()
        weights = self._grads[0].mean(dim=(1,2))
        cam = (weights[:,None,None]*self._feats[0]).sum(0)
        cam = F.relu(cam).cpu().numpy()
        return (cam-cam.min())/(cam.max()-cam.min()+1e-8)


def show_gradcam(model, df, n=6, thr=0.4, save=None, title="Grad-CAM"):
    cam_gen = GradCAM(model, model.backbone.blocks[-1][-1].conv_pwl)
    sample  = df.sample(n=min(n, len(df)), random_state=SEED)
    fig, axes = plt.subplots(2, n, figsize=(n*3, 6))
    fig.suptitle(title, fontsize=14, fontweight="bold")

    for i, (_, row) in enumerate(sample.iterrows()):
        try:    img_raw = read_img(str(row["image_path"]), size=IMG_SIZE)
        except: img_raw = np.zeros((IMG_SIZE,IMG_SIZE,3), dtype=np.uint8)
        aug = eval_aug(image=img_raw)["image"]
        try:
            cam = cv2.resize(cam_gen.generate(aug), (IMG_SIZE, IMG_SIZE))
            hm  = cv2.cvtColor(cv2.applyColorMap((cam*255).astype(np.uint8),
                               cv2.COLORMAP_JET), cv2.COLOR_BGR2RGB)
            ov  = np.clip(0.5*img_raw+0.5*hm, 0, 255).astype(np.uint8)
        except:
            ov  = img_raw
        with torch.no_grad():
            prob = torch.sigmoid(model(aug.unsqueeze(0).to(DEVICE))).item()
        tl = "Leukemia" if row["label"]==1 else "Normal"
        pl = "Leukemia" if prob>=thr else "Normal"
        cl = "green" if pl==tl else "red"
        axes[0,i].imshow(img_raw)
        axes[0,i].set_title(f"True:{tl}\nPred:{pl}\n{prob:.1%}", fontsize=9, color=cl)
        axes[0,i].axis("off")
        axes[1,i].imshow(ov)
        axes[1,i].set_title("Grad-CAM", fontsize=8)
        axes[1,i].axis("off")
    plt.tight_layout()
    if save: plt.savefig(save, dpi=150, bbox_inches="tight")
    plt.show(); print(f"  💾 {save}")

show_gradcam(model_b5, df_test, n=6, thr=opt_thr_b5,
             save=PLOTS_DIR/"gradcam_b5.png",
             title="Grad-CAM — EfficientNet-B5 Decisions")


# ┌─────────────────────────────────────────────────────────────────────────────┐
# │  CELL 14 — Full Inference Pipeline (Enhanced)                              │
# └─────────────────────────────────────────────────────────────────────────────┘

class LeukemiaDetectionSystem:
    """
    نظام كامل محسّن للكشف عن ابيضاض الدم.
    المراحل: Validation → Quality → Preprocessing → Ensemble+TTA → Calibrated Decision

    ★ الجديد:
    - Ensemble من نموذجين (B5 + B4)
    - TTA اختياري للحصول على أعلى دقة
    - Threshold محسوب تلقائياً من Validation Set
    - درجة ثقة مفصلة
    """
    VALID_THR  = 0.55
    UNC_LOW    = 0.30
    UNC_HIGH   = 0.60
    MIN_SHARP  = 80.0
    MIN_BRIGHT = 20
    MAX_BRIGHT = 240

    def __init__(self, validator, classifier_b5, classifier_b4,
                 device, ens_thr=0.40, use_tta=True):
        self.v       = validator.eval()
        self.c_b5    = classifier_b5.eval()
        self.c_b4    = classifier_b4.eval()
        self.dv      = device
        self.thr     = ens_thr
        self.use_tta = use_tta

    def _quality(self, img):
        gray   = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        sharp  = cv2.Laplacian(gray, cv2.CV_64F).var()
        brt    = float(gray.mean())
        issues = []
        if sharp < self.MIN_SHARP:   issues.append(f"Blurry (sharp={sharp:.0f})")
        if brt   < self.MIN_BRIGHT:  issues.append(f"Dark (brt={brt:.0f})")
        if brt   > self.MAX_BRIGHT:  issues.append(f"Overexposed (brt={brt:.0f})")
        return {"sharpness": sharp, "brightness": brt,
                "is_good": not issues, "issues": issues}

    @torch.no_grad()
    def predict(self, image_path: str) -> dict:
        res = {"path": image_path, "valid_prob": None, "quality": None,
               "prob": None, "pred": None, "conf": None, "msg": "", "warns": [],
               "prob_b5": None, "prob_b4": None}

        try:    img = read_img(str(image_path))
        except Exception as e:
            res["pred"] = "ERROR"; res["msg"] = f"❌ {e}"; return res

        # 1. Validation
        vt  = val_eval_aug(image=cv2.resize(img, (VAL_SIZE, VAL_SIZE)))["image"]
        vp  = torch.sigmoid(self.v(vt.unsqueeze(0).to(self.dv))).item()
        res["valid_prob"] = vp
        if vp < self.VALID_THR:
            res["pred"] = "INVALID"
            res["msg"]  = (f"❌ Image not valid (blood smear prob={vp:.1%})\n"
                           "   Please upload a real microscopic blood smear image.")
            return res

        # 2. Quality
        q = self._quality(img); res["quality"] = q
        for iss in q["issues"]: res["warns"].append(f"⚠️  {iss}")

        # 3. Preprocessing
        img_p = preprocess_blood(img)

        # 4. ★ Ensemble Prediction
        if self.use_tta:
            p_b5 = predict_tta(self.c_b5, img_p, self.dv)
            p_b4 = predict_tta(self.c_b4, img_p, self.dv)
        else:
            ct   = eval_aug(image=img_p)["image"]
            p_b5 = torch.sigmoid(self.c_b5(ct.unsqueeze(0).to(self.dv))).item()
            p_b4 = torch.sigmoid(self.c_b4(ct.unsqueeze(0).to(self.dv))).item()

        # متوسط مرجح: B5 يحصل على وزن أعلى (أحدث وأدق)
        prob = 0.60 * p_b5 + 0.40 * p_b4
        res["prob"] = prob; res["prob_b5"] = p_b5; res["prob_b4"] = p_b4

        # 5. ★ Calibrated Decision
        if self.UNC_LOW < prob < self.UNC_HIGH:
            res["pred"] = "UNCERTAIN"; res["conf"] = "LOW"
            res["msg"]  = (f"⚠️  Uncertain result ({prob:.1%}) — "
                           "recommend re-examination or specialist review.")
        elif prob >= self.thr:
            res["pred"] = "LEUKEMIA"
            res["conf"] = "HIGH" if prob >= 0.75 else "MEDIUM"
            res["msg"]  = (f"🔴 Leukemia suspected ({prob:.1%}) — "
                           "consult a hematologist immediately.")
        else:
            res["pred"] = "NORMAL"
            res["conf"] = "HIGH" if prob <= 0.15 else "MEDIUM"
            res["msg"]  = (f"🟢 Normal cells ({prob:.1%}) — "
                           "no signs of leukemia detected.")
        return res

    def show(self, r):
        print("\n" + "═"*60)
        print(f"  Image         : {Path(r['path']).name}")
        if r["valid_prob"]:
            print(f"  Blood Smear   : {r['valid_prob']:.1%}")
        if r["quality"]:
            q = r["quality"]
            print(f"  Quality       : {'✅ Good' if q['is_good'] else '⚠️  Issues'}"
                  f"  sharp={q['sharpness']:.0f}  bright={q['brightness']:.0f}")
        if r["prob_b5"] is not None:
            print(f"  B5 Prob       : {r['prob_b5']:.1%}")
            print(f"  B4 Prob       : {r['prob_b4']:.1%}")
            print(f"  Ensemble Prob : {r['prob']:.1%}")
        print(f"\n  {r['msg']}")
        for w in r["warns"]: print(f"  {w}")
        if r["conf"]:
            print(f"\n  Confidence    : {r['conf']}")
        print("═"*60)


# ★ إنشاء النظام مع الـ threshold المحسوب تلقائياً
system = LeukemiaDetectionSystem(
    validator        = val_model,
    classifier_b5    = model_b5,
    classifier_b4    = model_b4,
    device           = DEVICE,
    ens_thr          = ENS_THR,
    use_tta          = True,
)

print("\n" + "═"*65)
print("  🧪 Testing system on 6 samples (3 Normal + 3 Leukemia)")
print("═"*65)

# اختبار على عينات متوازنة
normal_samples   = df_test[df_test.label==0].sample(n=3, random_state=1)
leukemia_samples = df_test[df_test.label==1].sample(n=3, random_state=1)
test_samples     = pd.concat([normal_samples, leukemia_samples])

for _, row in test_samples.iterrows():
    r = system.predict(row["image_path"])
    true_label = "Normal" if row["label"] == 0 else "Leukemia"
    system.show(r)
    match = "✅ CORRECT" if (
        (r["pred"]=="NORMAL"   and row["label"]==0) or
        (r["pred"]=="LEUKEMIA" and row["label"]==1)
    ) else "❌ WRONG"
    print(f"  True Label: {true_label}  →  {match}")


# ┌─────────────────────────────────────────────────────────────────────────────┐
# │  CELL 15 — Final Summary                                                    │
# └─────────────────────────────────────────────────────────────────────────────┘

print("\n" + "═"*65)
print("  ✅ ENHANCED SYSTEM READY")
print("═"*65)
print(f"""
  📊 Dataset Summary:
  ─────────────────────────────────────────────
  C-NMC Leukemia               (primary)
  ALL 4-Class Cancer            (leukemia +)
  ★ BCCD Blood Cells            (normal +)
  ★ White Blood Cells Dataset   (normal +)
  ★ Peripheral Blood Cells      (normal +)
  ─────────────────────────────────────────────
  Total Images  : {len(df_leukemia):,}
  Normal   (0)  : {(df_leukemia.label==0).sum():,}
  Leukemia (1)  : {(df_leukemia.label==1).sum():,}

  🤖 Models:
  ─────────────────────────────────────────────
  • best_validator.pth   → MobileNetV3-Small
  • best_b5.pth          → EfficientNet-B5 + Attention
  • best_b4.pth          → EfficientNet-B4
  ─────────────────────────────────────────────

  🏆 Final Test Set Results (Ensemble + TTA):
  ─────────────────────────────────────────────
  ★ Recall (Sensitivity) : {test_m['recall']*100:.2f}%
  ★ AUC-ROC              : {test_m['auc']:.4f}
  ★ F2-Score             : {test_m['f2']*100:.2f}%
    Specificity          : {test_m['specificity']*100:.2f}%
    Accuracy             : {test_m['accuracy']*100:.2f}%
    False Negatives      : {test_m['fn']}  ← closer to zero = better
  ─────────────────────────────────────────────

  📁 Models : {MODELS_DIR}
  📊 Plots  : {PLOTS_DIR}
""")
```
